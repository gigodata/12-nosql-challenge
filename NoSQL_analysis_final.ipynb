### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

# NB: These instructions are ambiguous and contradictory!!!

## 'Top 5' infers the highest value--the question is: Top 5 of (what??) -- every 'RatingValue' == 5, so there is no inherent highest value, if all RatingValue are equal. There is no differentiation, so 'Top 5' is impossible.

## Sorting by lowest hygiene score indicates that least restaurants with hygienic scores will be at the top of the returned results set.

## Given these conflicting prioritisations, the returned results set are sorted low to high in terms of the Hygiene score.

```python
# Retrieve location data for 'Penang Flavours'
penang = establishments.find_one({"BusinessName": "Penang Flavours"}, {"geocode": 1})
```

```python
# Set degree of search distance from "Penang Flavours"
degree_search = 0.01
```

```python
# Retrieve longitude & latitude for 'Penang Flavours'
longitude = penang['geocode']['longitude']
latitude = penang['geocode']['latitude']
```

```python
# Restrict to documents where RatingValue == 5
# Search within 0.01 degree on either side of the latitude and longitude.
query2 = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}
```

```python
# Sort by hygiene score in ascending order, lowest to highest
sort = [("scores.Hygiene", 1)]
```

```python
# Return documents that meet query criteria, sorted according to definition, and limited to ambiguously defined Top 5.
result2 = list(establishments.find(query2).sort(sort).limit(5))
```

```python
# Print the results
pprint(result2)
```

```python
# Convert the result to a Pandas DataFrame
df2 = pd.DataFrame(result2)
df2
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that will: 

# 1. Match establishments with a hygiene score of 0
match = {"$match": {"scores.Hygiene": 0}}
```

```python
# 2. Group the matches by Local Authority
group = {"$group": {"_id": "$LocalAuthorityName", "total": {"$sum": 1}}}
```

```python
# 3. Sort the matches from highest to lowest
sort = {"$sort": {"total": -1}}
```

```python
# Set pipeline criteria
pipeline = [match, group, sort]
```

```python
# Create aggregate pipeline
result3 = list(establishments.aggregate(pipeline))
```

```python
# Print the number of documents in the result
pprint(len(result3))
```

```python
# Print the first 10 results
pprint(result3[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df3 = pd.DataFrame(result3)
```

```python
# Rename _ID to LocalAuthorityName for clarity
df3.rename(columns={'_id': 'LocalAuthorityName'}, inplace = True)
```

```python
# Display the number of rows in the DataFrame
len(df3)
```

```python
# Display the first 10 rows of the DataFrame
df3.head(10)
```
